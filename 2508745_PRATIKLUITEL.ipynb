Task - 3 - Outlier Detection:
Tasks:
1. Identify outlier countries in South Asia based on their Score and GDP per Capita

```python
Q1 = df_sasia['Log GDP per capita'].quantile(0.25)
Q3 = df_sasia['Log GDP per capita'].quantile(0.75)
IQR = Q3 - Q1

lower_gdp = Q1 - 1.5*IQR
upper_gdp = Q3 + 1.5*IQR
```

2. Define outliers using the 1.5  IQR rule

```python
outliers_gdp = df_sasia[(df_sasia['Log GDP per capita'] < lower_gdp) |
                        (df_sasia['Log GDP per capita'] > upper_gdp)]
outliers_gdp
```

3. Create a scatter plot with GDP per Capita on the x-axis and Score on the y-axis, highlighting outliers
in a different color

```python
plt.figure(figsize=(8,6))
plt.scatter(df_sasia['Log GDP per capita'], df_sasia['score'], label='Normal')

# Mark outliers in red
plt.scatter(outliers_gdp['Log GDP per capita'], outliers_gdp['score'], color='red', label='Outliers')

plt.xlabel("GDP per Capita")
plt.ylabel("Score")
plt.title("South Asia Outliers based on Score vs GDP")
plt.legend()
plt.show()
```

4. Discuss the characteristics of these outliers and their potential impact on regional averages

Answer
=>  Afghanistan stands as the most pronounced negative outlier across the South Asian dataset, with its happiness score much lower than the rest of South Asia owing to appallingly disastrous socio-economic conditions, ongoing political instability, and absence of basic public goods. This extreme low value been pulling the regional average score down, giving an impression of a South Asia that is less happy than most of its countries actually are. Hence, Afghanistan widens the data spread (by range and variance) such that South Asia appears to have well-being weaker than what the median portrays.

The more Maldives acts as a positive outlier with regard to economic dimensions since it has better GDP, healthcare services, and economy buoyed by tourism. This positive effect upside biasing the average score is in itself a systematic distortion of the true central tendency for the region. Bhutan and Sri Lanka also enjoy a somewhat higher standing with respect to the quality of environment and available social support, albeit nowhere near as extreme as that of the Maldives.

As seen, the above-discussed outliers on both ends, extremely low and moderately high, are likely causing gross distortions on the summary statistics of the region. This also makes direct comparison tedious, which in turn prompts misinterpretation of the real happiness level of South Asia.

Complete the following task:

1.	Descriptive Statistics:

	Calculate the mean, Standard deviation of the score for both South Asia and Middle East.

  Which region has higher happiness Scores on average?


```python
mean_sasia = df_sasia['score'].mean()
std_sasia = df_sasia['score'].std()

mean_me = df_me['score'].mean()
std_me = df_me['score'].std()

print("South Asia Mean:", mean_sasia)
print("South Asia STD:", std_sasia)
print("Middle East Mean:", mean_me)
print("Middle East STD:", std_me)

if mean_me > mean_sasia:
    print("Middle East is happier on average.")
else:
    print("South Asia is happier on average.")
```

3.	Metric Comparisons:
	Compare key metrics like GDP per Capita, Social Support, and Healthy Life Expectancy between the regions using grouped bar charts.
	Which metrics show the largest disparity between the two regions?


```python
metrics = ["Log GDP per capita", "Social support", "Healthy life expectancy"]

df_metric_compare = pd.DataFrame({
    "Metric": metrics,
    "South Asia": [df_sasia[m].mean() for m in metrics],
    "Middle East": [df_me[m].mean() for m in metrics]
})
df_metric_compare
```

```python
x = range(len(metrics))
plt.figure(figsize=(8,5))
plt.bar([p - 0.2 for p in x], df_metric_compare["South Asia"], width=0.4, label="South Asia")
plt.bar([p + 0.2 for p in x], df_metric_compare["Middle East"], width=0.4, label="Middle East")
plt.xticks(x, metrics)
plt.ylabel("Average Value")
plt.title("Metric Comparisons")
plt.legend()
plt.show()
```

4.	Happiness Disparity:

    Compute the range (max - min) and coefficient of variation (CV) for Score in both regions.
    
    Which region has greater variability in happiness?


```python
range_sasia = df_sasia['score'].max() - df_sasia['score'].min()
range_me = df_me['score'].max() - df_me['score'].min()

cv_sasia = std_sasia / mean_sasia
cv_me = std_me / mean_me

print("South Asia Range:", range_sasia)
print("Middle East Range:", range_me)
print("South Asia CV:", cv_sasia)
print("Middle East CV:", cv_me)

if range_me > range_sasia:
    print("Middle East has more variability.")
else:
    print("South Asia has more variability.")
```
